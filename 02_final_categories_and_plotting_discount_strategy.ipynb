* Revenue peaks at 5–10% discount — then declines sharply

There is no positive relationship between higher discounts and higher spending.

# 5. Do top products actually need discounts?

### “The analysis shows that top-performing products generate the majority of their revenue under discounted conditions. However, given that previous analyses indicate discounts do not increase purchase volume or revenue per order, this suggests that discounts are being applied to products with strong inherent demand, leading to unnecessary revenue loss.”

### “Top products appear to be heavily discounted, yet earlier findings show no evidence of increased demand due to discounting. This indicates that discounts are likely reducing revenue on already high-performing products rather than driving additional sales.”

### “We are discounting our best-selling products — which is the least efficient place to apply discounts.”


```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

df = final_df_clean.copy()

# 1) Base columns
df["discounted_revenue"] = df["unit_price"] * df["product_quantity"]
df["is_discounted"] = df["discount_pct"] > 0

# 2) Top 10 SKUs by discounted revenue
top_skus = (
    df.groupby("sku")["discounted_revenue"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .index
)

top_df = df[df["sku"].isin(top_skus)].copy()

# 3) Clean sku -> name map
sku_name_map = (
    df.groupby("sku", as_index=False)["name"]
    .first()
)

# 4) Aggregate revenue by sku + discount status
comparison = (
    top_df.groupby(["sku", "is_discounted"], as_index=False)["discounted_revenue"]
    .sum()
)

# 5) Pivot safely
pivot_df = comparison.pivot(
    index="sku",
    columns="is_discounted",
    values="discounted_revenue"
).fillna(0)

# 6) Rename safely without assuming both columns are present
pivot_df = pivot_df.rename(columns={False: "No Discount", True: "Discounted"})

if "No Discount" not in pivot_df.columns:
    pivot_df["No Discount"] = 0
if "Discounted" not in pivot_df.columns:
    pivot_df["Discounted"] = 0

pivot_df = pivot_df[["No Discount", "Discounted"]].reset_index()

# 7) Add names
pivot_df = pivot_df.merge(sku_name_map, on="sku", how="left")

# 8) Short labels
pivot_df["short_name"] = pivot_df["name"].apply(
    lambda x: x[:40] + "..." if isinstance(x, str) and len(x) > 40 else str(x)
)

pivot_df["label"] = pivot_df["short_name"] + " (" + pivot_df["sku"].astype(str) + ")"

plot_df = pivot_df.set_index("label")[["No Discount", "Discounted"]]

# 9) Plot
ax = plot_df.plot(kind="barh", figsize=(12, 8))

plt.xlabel("Total Product Revenue")
plt.ylabel("Product")
plt.title("Top 10 Products: Discounted vs Non-Discounted Revenue")

plt.gca().xaxis.set_major_formatter(
    mticker.FuncFormatter(lambda x, _: f"€{x:,.0f}")
)

plt.tight_layout()
plt.show()
```
